```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
five_factor = pd.read_csv('F-F_Research_Data_5_Factors_2x3.csv', skiprows=3)
five_factor.columns = ['date', 'MKT-RF', 'SMB', 'HML', 'RMW', 'CMA', 'RF']
five_factor = five_factor[0:699]
five_factor = five_factor.astype(float)
five_factor['date']=pd.to_datetime(five_factor["date"], format="%Y%m")
five_factor.set_index('date', inplace = True)
five_factor.head()
```

### Table 1

```python
#Table 1 - Regressions
#regression on mkt-rf
Y = five_factor['MKT-RF']
X = five_factor[['SMB', 'HML', 'RMW', 'CMA']]
X2 = sm.add_constant(X)
mkt = sm.OLS(Y, X2).fit()
mkt.summary2().tables[1]["Coef."]
mkt.summary2().tables[1]["t"]
mkt.summary2().tables[0][1][6]

#regression on smb
Y = five_factor['SMB']
X = five_factor[['MKT-RF', 'HML', 'RMW', 'CMA']]
X2 = sm.add_constant(X)
smb = sm.OLS(Y, X2).fit()
smb.summary2().tables[1]["Coef."]
smb.summary2().tables[1]["t"]
smb.summary2().tables[0][1][6]

#regression on hml
Y = five_factor['HML']
X = five_factor[['MKT-RF', 'SMB', 'RMW', 'CMA']]
X2 = sm.add_constant(X)
hml = sm.OLS(Y, X2).fit()
hml.summary2().tables[1]["Coef."]
hml.summary2().tables[1]["t"]
hml.summary2().tables[0][1][6]

#regression on rmw
Y = five_factor['RMW']
X = five_factor[['MKT-RF', 'SMB', 'HML', 'CMA']]
X2 = sm.add_constant(X)
rmw = sm.OLS(Y, X2).fit()
rmw.summary2().tables[1]["Coef."]
rmw.summary2().tables[1]["t"]
rmw.summary2().tables[0][1][6]

#regression on rmw
Y = five_factor['CMA']
X = five_factor[['MKT-RF', 'SMB', 'HML', 'RMW']]
X2 = sm.add_constant(X)
cma = sm.OLS(Y, X2).fit()
cma.summary2().tables[1]["Coef."]
cma.summary2().tables[1]["t"]
cma.summary2().tables[0][1][6]
```

```python
mktv=mkt.summary2().tables[1]["Coef."]
mktv['MKT-RF']=""
mktv['R-squared']=mkt.summary2().tables[0][1][6]
mktt=mkt.summary2().tables[1]["t"]
mktt['MKT-RF']=""
mktt['R-squared']=""
mktd=pd.merge(pd.DataFrame(mktv),pd.DataFrame(mktt), on=mktv.index)
mktd.columns=['key','MKT-RF','t(mkt-rf)']
mktd=mktd.T
mktd=mktd.rename(columns=mktd.iloc[0], inplace=False)
mktd=mktd[['const','MKT-RF', 'SMB', 'HML', 'RMW', 'CMA', 'R-squared']]
mktd=mktd.iloc[1: , :]

smbv=smb.summary2().tables[1]["Coef."]
smbv['SMB']=""
smbv['R-squared']=smb.summary2().tables[0][1][6]
smbt=smb.summary2().tables[1]["t"]
smbt['SMB']=""
smbt['R-squared']=""
smbd=pd.merge(pd.DataFrame(smbv),pd.DataFrame(smbt), on=smbv.index)
smbd.columns=['key','SMB','t(SMB)']
smbd=smbd.T
smbd=smbd.rename(columns=smbd.iloc[0], inplace=False)
smbd=smbd[['const','MKT-RF', 'SMB', 'HML', 'RMW', 'CMA', 'R-squared']]
smbd=smbd.iloc[1: , :]

hmlv=hml.summary2().tables[1]["Coef."]
hmlv['HML']=""
hmlv['R-squared']=hml.summary2().tables[0][1][6]
hmlt=hml.summary2().tables[1]["t"]
hmlt['HML']=""
hmlt['R-squared']=""
hmld=pd.merge(pd.DataFrame(hmlv),pd.DataFrame(hmlt), on=hmlv.index)
hmld.columns=['key','HML','t(HML)']
hmld=hmld.T
hmld=hmld.rename(columns=hmld.iloc[0], inplace=False)
hmld=hmld[['const','MKT-RF', 'SMB', 'HML', 'RMW', 'CMA', 'R-squared']]
hmld=hmld.iloc[1: , :]

rmwv=rmw.summary2().tables[1]["Coef."]
rmwv['RMW']=""
rmwv['R-squared']=rmw.summary2().tables[0][1][6]
rmwt=rmw.summary2().tables[1]["t"]
rmwt['RMW']=""
rmwt['R-squared']=""
rmwd=pd.merge(pd.DataFrame(rmwv),pd.DataFrame(rmwt), on=rmwv.index)
rmwd.columns=['key','RMW','t(RMW)']
rmwd=rmwd.T
rmwd=rmwd.rename(columns=rmwd.iloc[0], inplace=False)
rmwd=rmwd[['const','MKT-RF', 'SMB', 'HML', 'RMW', 'CMA', 'R-squared']]
rmwd=rmwd.iloc[1: , :]

cmav = cma.summary2().tables[1]["Coef."]
cmav['CMA']=""
cmav['R-squared']=cma.summary2().tables[0][1][6]
cmat=cma.summary2().tables[1]["t"]
cmat['CMA']=""
cmat['R-squared']=""
cmad=pd.merge(pd.DataFrame(cmav),pd.DataFrame(cmat), on=cmav.index)
cmad.columns=['key','CMA','t(CMA)']
cmad=cmad.T
cmad=cmad.rename(columns=cmad.iloc[0], inplace=False)
cmad=cmad[['const','MKT-RF', 'SMB', 'HML', 'RMW', 'CMA', 'R-squared']]
cmad=cmad.iloc[1: , :]

Table1 = pd.concat([mktd,smbd,hmld,rmwd,cmad])
Table1
```
